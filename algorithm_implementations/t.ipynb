```python
import gymnasium as gym
import random

env = gym.make("CartPole-v1", render_mode="human")

episodes = 10
for episode in range(episodes):
    state, info = env.reset()
    done = False
    score = 0
    while not done:
        action = random.choice([0, 1])
        state, reward, done, truncated, info = env.step(action) #done signifies if the episode is over
        score += reward
        env.render()
    print(f"Episode {episode + 1}: Score = {score}")
env.close()
# Note: The above code will open a window to render the CartPole environment.
# Make sure to close the window after running the code.
# The code uses the OpenAI Gym library to create a simple reinforcement learning environment.
# The CartPole environment is a classic control problem where the goal is to balance a pole on a cart.
# The agent takes random actions and the score is the total reward received in each episode.
# The code runs for 10 episodes, rendering the environment and printing the score for each episode.
# The environment is closed at the end to free up resources.
# Note: The code uses the `gymnasium` package, which is a fork of OpenAI Gym.
```

**Implementing different algorithms from the Sutton and Barto Book on testing environments FrozenLake, Cliffwalking, Taxi-v3, CartPole.**

*Dynamic Programming Policy Evaluation using CliffWalking environment.*

```python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class PolicyEvaluator:
    """
    Implements Iterative Policy Evaluation (Sutton & Barto, Ch. 4) 
    for a fixed policy in a discrete, finite MDP using Dynamic Programming.
    """

    def __init__(self, env, gamma=0.9, theta=1e-6):
        self.env = env
        self.gamma = gamma  # Discount factor (how far-sighted the agent is)
        self.theta = theta  # Convergence threshold
        self.V = np.zeros(env.observation_space.n)  # Initial value function

    def evaluate_policy(self, policy, max_iterations=1000, verbose=True):
        """
        Evaluate the given policy using iterative policy evaluation:
        V(s) <- sum_a pi(a|s) sum_{s', r} p(s', r | s, a) [r + gamma * V(s')]
        """
        self.V = np.zeros(self.env.observation_space.n)
        for iteration in range(max_iterations):
            delta = 0  # Tracks maximum change for convergence check
            old_V = self.V.copy()
            for state in range(self.env.observation_space.n):
                v = 0
                # Get the action probabilities for the current state
                action_probs = policy.get(state, {a: 0.25 for a in range(self.env.action_space.n)})
                for action, prob in action_probs.items():
                    # env.P[state][action] gives list of (probability, next_state, reward, done)
                    for (transition_prob, next_state, reward, done) in self.env.P[state][action]:
                        v += prob * transition_prob * (reward + self.gamma * self.V[next_state])
                self.V[state] = v
                delta = max(delta, abs(old_V[state] - self.V[state]))
            if verbose and iteration % 10 == 0:
                print(f"Iteration {iteration}: Max delta = {delta:.6f}")
            if delta < self.theta:
                if verbose:
                    print(f"Converged after {iteration + 1} iterations")
                return self.V, iteration + 1
        print(f"Did not converge after {max_iterations} iterations")
        return self.V, max_iterations

    def visualize_value_function(self, title="Value Function"):
        """
        Display value function as a heatmap.
        """
        V_grid = self.V.reshape((4, 12))  # CliffWalking is a 4x12 grid
        plt.figure(figsize=(12, 4))
        sns.heatmap(V_grid, annot=True, fmt=".1f", cmap="RdYlBu_r", cbar_kws={'label': 'Value'})
        plt.title(title)
        plt.xlabel('Column')
        plt.ylabel('Row')
        # Mark START, GOAL, and CLIFF regions
        plt.text(0.5, 3.5, 'START', ha='center', va='center', 
                 bbox=dict(boxstyle="round,pad=0.3", facecolor="green", alpha=0.7))
        plt.text(11.5, 3.5, 'GOAL', ha='center', va='center',
                 bbox=dict(boxstyle="round,pad=0.3", facecolor="gold", alpha=0.7))
        for col in range(1, 11):
            plt.text(col + 0.5, 3.5, 'CLIFF', ha='center', va='center',
                     bbox=dict(boxstyle="round,pad=0.2", facecolor="red", alpha=0.7))
        plt.tight_layout()
        plt.show()

def create_random_policy(env):
    """
    Assign equal probability to all actions for each state (uniform random policy).
    """
    return {state: {a: 1 / env.action_space.n for a in range(env.action_space.n)} for state in range(env.observation_space.n)}

def create_optimal_policy(env):
    """
    A manually designed 'safe' policy that avoids the cliff.
    Right if above goal row, Up otherwise.
    """
    policy = {}
    for state in range(env.observation_space.n):
        row, col = divmod(state, 12)
        action_probs = {a: 0.0 for a in range(env.action_space.n)}
        if state == 47:  # Goal
            action_probs = {a: 0.25 for a in range(env.action_space.n)}
        elif row < 3:
            if col < 11:
                action_probs[1] = 1.0  # Right
            else:
                action_probs[2] = 1.0  # Down
        elif row == 3:
            if col == 0:
                action_probs[0] = 1.0  # Up
            elif col == 11:
                action_probs = {a: 0.25 for a in range(env.action_space.n)}
            else:
                action_probs[0] = 1.0  # Up
        policy[state] = action_probs
    return policy

def create_risky_policy(env):
    """
    A policy that risks walking next to the cliff to reach the goal faster.
    """
    policy = {}
    for state in range(env.observation_space.n):
        row, col = divmod(state, 12)
        action_probs = {a: 0.0 for a in range(env.action_space.n)}
        if state == 47 or (row == 3 and 1 <= col <= 10):
            action_probs = {a: 0.25 for a in range(env.action_space.n)}
        elif row == 3 and col < 11:
            action_probs[1] = 1.0  # Right
        elif row < 3:
            action_probs[2] = 1.0  # Down
        else:
            action_probs[1] = 1.0  # Right
        policy[state] = action_probs
    return policy

if __name__ == "__main__":
    env = gym.make("CliffWalking-v0")
    env = env.unwrapped  # Unwrap to access transition probabilities (.P)
    evaluator = PolicyEvaluator(env, gamma=0.9)

    # Define multiple policies to evaluate
    policies = {
        "Random Policy": create_random_policy(env),
        "Safe Policy (Optimal)": create_optimal_policy(env),
        "Risky Policy": create_risky_policy(env)
    }

    # Evaluate all policies and visualize
    results = {}
    for name, policy in policies.items():
        print(f"\nEvaluating {name}...")
        V, iters = evaluator.evaluate_policy(policy, verbose=False)
        results[name] = (V.copy(), iters)
        print(f"Converged in {iters} iterations | Start value: {V[36]:.2f}")

    for name, (V, _) in results.items():
        evaluator.V = V
        evaluator.visualize_value_function(f"{name} - Value Function")

    # Summary of results
    print("\n=== Policy Comparison Summary ===")
    for name, (V, iters) in results.items():
        print(f"{name:24}: Start Value = {V[36]:7.3f}, Iterations = {iters:3d}")
```
